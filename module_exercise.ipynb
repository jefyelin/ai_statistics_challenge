```python
import pandas as pd
import matplotlib.pyplot as plt

dict_invoicing = {
    "date_ref": [
        "2023-01-01",
        "2020-02-01",
        "2021-03-01",
        "2022-04-01",
        "2023-05-01",
        "2023-06-01",
        "2020-07-01",
        "2021-08-01",
        "2022-09-01",
        "2023-10-01",
        "2022-11-01",
        "2023-12-01",
    ],
    "value": [
        400000,
        890000,
        760000,
        430000,
        920000,
        340000,
        800000,
        500000,
        200000,
        900000,
        570000,
        995000,
    ],
}
```

```python
data_frame = pd.DataFrame(dict_invoicing)
data_frame["date_ref"] = pd.to_datetime(data_frame["date_ref"])
data_frame["month"] = data_frame["date_ref"].dt.month

data_frame
```

### Mean

```python
mean = float(round(data_frame["value"].mean(), 2))

mean
```

### Vertical Bar

```python
vertical_bar = data_frame.plot.bar(x="month", y="value", legend=False)

vertical_bar.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{int(x):,}'))
vertical_bar.set_xlabel("Date Reference")
vertical_bar.set_ylabel("Value")
vertical_bar.set_title("Invoicing Over Time")
vertical_bar.xaxis.set_tick_params(rotation=0)

plt.show()
```

## Line

```python
line = data_frame.plot.line(x="month", y="value")

line.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{int(x):,}'))

line.set_xlabel("Date Reference")
line.set_ylabel("Value")
line.set_title("Invoicing Over Time")
line.xaxis.set_tick_params(rotation=0)

plt.show()
```
